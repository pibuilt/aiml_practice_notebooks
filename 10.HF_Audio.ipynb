```python
from transformers import pipeline
from IPython.display import Audio, display
import tempfile


tts = pipeline("text-to-audio")
text = "Hello! This is a text-to-speech demo played directly in the notebook."
audio_output = tts(text)

print(audio_output.keys())
```

```python
display(Audio(audio_output["audio"], rate=audio_output["sampling_rate"], autoplay=True))
```
