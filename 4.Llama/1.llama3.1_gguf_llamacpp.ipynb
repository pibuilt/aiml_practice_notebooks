```python
from huggingface_hub import hf_hub_download

model_name = "dphn/Dolphin3.0-Llama3.1-8B-GGUF"
model_file = "Dolphin3.0-Llama3.1-8B-Q5_K_M.gguf"

model_path = hf_hub_download(
    repo_id=model_name,
    filename=model_file
)
```

```python
from llama_cpp import Llama

llm = Llama(
    model_path=model_path,
    n_threads=2,
    n_batch=512,
    n_gpu_layers=35
)
```

```python
prompt = "Explain gradient descent in simple terms"

prompt_template = f"""
SYSTEM: You are a helpful AI assistant.

USER: {prompt}

ASSISTANT:
"""
```

```python
response = llm(
    prompt=prompt_template,
    max_tokens=256,
    temperature=0.7,
    top_p=0.95,
    repeat_penalty=1.1
)
```

```python
print(response["choices"][0]["text"])
```

```python
questions = [
    "Write Python code for linear regression",
    "Explain transformers in deep learning",
    "What is overfitting?"
]

for q in questions:
    prompt_template = f"""
    SYSTEM: You are a helpful AI assistant.
    USER: {q}
    ASSISTANT:
    """

    res = llm(prompt=prompt_template, max_tokens=200)
    print("\nQ:", q)
    print(res["choices"][0]["text"])
```
