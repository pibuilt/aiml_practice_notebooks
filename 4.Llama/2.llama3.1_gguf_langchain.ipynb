```python
from huggingface_hub import hf_hub_download

model_name = "dphn/Dolphin3.0-Llama3.1-8B-GGUF"
model_file = "Dolphin3.0-Llama3.1-8B-Q5_K_M.gguf"

model_path = hf_hub_download(
    repo_id=model_name,
    filename=model_file
)
```

```python
from llama_cpp import Llama

llm_cpp = Llama(
    model_path=model_path,
    n_gpu_layers=20,
    n_batch=512,
    n_threads=2
)
```

```python
from langchain_community.llms import LlamaCpp
```

```python
llm = LlamaCpp(
    model_path=model_path,
    n_gpu_layers=35,
    n_batch=512,
    n_ctx=4096,
    temperature=0.7,
    verbose=True
)
```

```python
response = llm.invoke("Explain gradient descent simply")
print(response)
```

```python
from langchain_core.prompts import PromptTemplate

template = """
You are a helpful AI assistant.

Question: {question}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["question"],
    template=template
)

final_prompt = prompt.format(question="What is overfitting?")
print(llm.invoke(final_prompt))
```

```python
from langchain_classic.chains import LLMChain

chain = LLMChain(llm=llm, prompt=prompt)

print(chain.invoke("Explain transformers in deep learning"))
```
