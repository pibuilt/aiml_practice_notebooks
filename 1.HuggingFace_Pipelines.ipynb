## 1. text-generation

```python
# Create a text generation pipeline
generator = pipeline("text-generation")

# Show the default model and tokenizer used
print("Default model:", generator.model.name_or_path)
print("Default tokenizer:", generator.tokenizer.name_or_path)

# Show default generation config
print("Default generation config:", generator.model.config)

# Generate text from a prompt
prompt = "Once upon a time in a futuristic city,"
result = generator(prompt, max_length=50)
print("\nGenerated text:")
print(result[0]['generated_text'])
```

```python
# Create a text generation pipeline using Qwen3
generator = pipeline(
    "text-generation",
    model="Qwen/Qwen3-0.6B",  # The custom Qwen3 model
    device=0,                  # GPU index (0 for first GPU), remove if using CPU
    temperature=0.6,           # Controls randomness (0 = deterministic, higher = more creative)
    top_k=20,                  # Top-K sampling: only consider top 20 tokens at each step
    top_p=0.95,                # Top-p (nucleus) sampling: consider tokens covering 95% of total probability mass
                               # Probability mass = cumulative likelihood of the most probable tokens
    do_sample=True,            # Enable sampling (required for temperature/top-k/top-p to have effect)
    max_new_tokens=200         # Maximum number of tokens to generate
)

prompt = "Artificial intelligence will"
result = generator(prompt)
print(result[0]['generated_text'])
```
