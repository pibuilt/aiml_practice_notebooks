## 1. text-generation

```python
# Create a text generation pipeline
generator = pipeline("text-generation")

# Show the default model and tokenizer used
print("Default model:", generator.model.name_or_path)
print("Default tokenizer:", generator.tokenizer.name_or_path)

# Show default generation config
print("Default generation config:", generator.model.config)

# Generate text from a prompt
prompt = "Once upon a time in a futuristic city,"
result = generator(prompt, max_length=50)
print("\nGenerated text:")
print(result[0]['generated_text'])
```

```python
# Create a text generation pipeline using Qwen3
generator = pipeline(
    "text-generation",
    model="Qwen/Qwen3-0.6B",  # The custom Qwen3 model
    device=0,                  # GPU index (0 for first GPU), remove if using CPU
    temperature=0.6,           # Controls randomness (0 = deterministic, higher = more creative)
    top_k=20,                  # Top-K sampling: only consider top 20 tokens at each step
    top_p=0.95,                # Top-p (nucleus) sampling: consider tokens covering 95% of total probability mass
                               # Probability mass = cumulative likelihood of the most probable tokens
    do_sample=True,            # Enable sampling (required for temperature/top-k/top-p to have effect)
    max_new_tokens=200         # Maximum number of tokens to generate
)

prompt = "Artificial intelligence will"
result = generator(prompt)
print(result[0]['generated_text'])
```

## 2. text-to-audio

```python
from IPython.display import Audio, display
import tempfile
```

```python
# Create text-to-audio pipeline
tts = pipeline("text-to-audio")

# Text to convert
text = "Hello! This is a text-to-speech demo played directly in the notebook."

# Generate audio
audio_output = tts(text)

# Check output keys (should contain 'waveform' and 'sampling_rate')
print(audio_output.keys())

# Play inline in notebook
display(Audio(audio_output["audio"], rate=audio_output["sampling_rate"], autoplay=True))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 1️⃣ Load model & tokenizer
model_name = "Qwen/Qwen3-0.6B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# 2️⃣ Prepare input text
text = "Hello world! This is a test for input IDs."

# 3️⃣ Tokenize (convert text → token IDs)
inputs = tokenizer(text, return_tensors="pt")  # returns a dict with input_ids, attention_mask
input_ids = inputs["input_ids"]

# 4️⃣ Print input_ids & shape
print("Input IDs:", input_ids)
print("Shape of input_ids:", input_ids.shape)

# 5️⃣ Optional: pass through model
with torch.no_grad():
    outputs = model(**inputs)
    # logits shape: (batch_size, sequence_length, vocab_size)
    print("Logits shape:", outputs.logits.shape)
```
