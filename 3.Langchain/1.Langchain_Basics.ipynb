```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=100,
    temperature=0.7,
    device_map="auto"
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
def chat_prompt(user_input):
    return f"<|user|>\n{user_input}\n<|assistant|>"
```

```python
llm.invoke(chat_prompt("Explain embeddings in simple terms"))
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["topic"],
    template="<|user|>\nExplain {topic} in simple terms\n<|assistant|>"
)

llm.invoke(prompt.format(topic="Batman"))
```

```python
from langchain_classic.chains import LLMChain

chain = LLMChain(llm=llm, prompt=prompt)

chain.run("RAG systems in AIML")
```

```python
from langchain_classic.chains import SimpleSequentialChain

prompt1 = PromptTemplate(
    input_variables=["topic"],
    template="<|user|>\nGive a short explanation of {topic}\n<|assistant|>"
)

prompt2 = PromptTemplate(
    input_variables=["text"],
    template="<|user|>\nSummarize this:\n{text}\n<|assistant|>"
)

chain1 = LLMChain(llm=llm, prompt=prompt1)
chain2 = LLMChain(llm=llm, prompt=prompt2)

seq_chain = SimpleSequentialChain(chains=[chain1, chain2])

seq_chain.run("embeddings")
```

```python
from langchain_classic.chains import SequentialChain

prompt1 = PromptTemplate(
    input_variables=["topic"],
    output_key="explanation",
    template="<|user|>\nExplain {topic}\n<|assistant|>"
)

prompt2 = PromptTemplate(
    input_variables=["explanation"],
    output_key="summary",
    template="<|user|>\nSummarize:\n{explanation}\n<|assistant|>"
)

chain1 = LLMChain(llm=llm, prompt=prompt1, output_key="explanation")
chain2 = LLMChain(llm=llm, prompt=prompt2, output_key="summary")

multi_chain = SequentialChain(
    chains=[chain1, chain2],
    input_variables=["topic"],
    output_variables=["explanation", "summary"]
)

multi_chain({"topic": "vector databases"})
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("1.langchaintext.txt")
documents = loader.load()
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)

docs = splitter.split_documents(documents)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)
```

```python
retriever = db.as_retriever()

docs = retriever.invoke("What is this document about?")
```

```python
rag_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
<|user|>
Answer ONLY using the context.

Context:
{context}

Question:
{question}

Keep answer short.
<|assistant|>
"""
)
```

```python
docs = retriever.invoke("What is LangChain?")

context = "\n".join([doc.page_content for doc in docs])

final_prompt = rag_prompt.format(
    context=context,
    question="What is LangChain?"
)

llm.invoke(final_prompt)
```

```python
from langchain_classic.chains import RetrievalQA

qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)

qa.run("Explain LangChain")
```
