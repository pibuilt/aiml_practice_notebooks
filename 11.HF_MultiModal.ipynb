```python
from transformers import pipeline

# Load the small multimodal model
any_to_any = pipeline(
    task="any-to-any",
    model="HuggingFaceTB/SmolVLM-256M-Instruct"
)

inputs = {
    "image": r"C:\Users\Piyush\Pictures\tmp3c2hh01f.png",  # your local image
    "text": "What is in the image."
}

output = any_to_any(inputs, max_new_tokens=50)
print(output)
```
