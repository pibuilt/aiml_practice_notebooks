```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "Qwen/Qwen3-0.6B"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",     # automatically use best dtype (fp16/bf16) for your device
    device_map="auto"       # automatically place model layers on available GPUs/CPU
)
```

```python
prompt = "Explain the importance of Hugging Face AutoModel."
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
print(inputs)
print(inputs["input_ids"])
print(inputs["input_ids"].shape)

outputs = model.generate(
    **inputs,
    max_new_tokens=150,     # limit the response length
    temperature=0.6,        # randomness in generation
    top_p=0.95,             # nucleus sampling
    top_k=20                # limit to top-k tokens
)
print(outputs)
```

```python
text_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(text_output)
```
